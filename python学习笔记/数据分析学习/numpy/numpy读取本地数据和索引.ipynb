# numpy索引和切片

```python
b = np.arange(50).reshape((10,5))
print(b)
print("#"*100)

#取行
print(b[0])
print("#"*100)


#取连续多行
print(b[0:3])
print("#"*100)

#取不连续多行
print(b[[2,4,8]])
print("#"*100)

#取列
print(b[:,0])
print("#"*100)


#取连续多列
print(b[:,2:5])
print("#"*100)

#取不连续多列
print(b[:,[0,2,3]])
print("#"*100)

#取3到5行，2到4列的值
print(b[2:5,1:4])
print("#"*100)

#取多个不相邻的点（0,0）（2,1）
print(b[[0,2],[0,1]])
```

# numpy对数值的修改

```python
#将小于10的值改为0
b[b<10] = 0
print(b)
print("#"*100)


np.where(b<25,100,300)


#b.clip(10,18)将小于等于10的替换为10，大于等于18的替换为18

```
