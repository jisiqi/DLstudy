```python
import numpy as np 
import pandas as pd
```

```python
randn = np.random.randn(8,5)
a = np.arange(1,9,1)
df = pd.DataFrame(randn,index=a,columns=list("abcde"))
```

```python
#打印出首尾行数，默认为5


print(df.head())
print("#"*100)
print(df.tail())
```

```python
#行列转换
df.T
```

```python
print(df.sort_values(by = 'b',ascending=True))
print("#"*100)
print(df.sort_values(by = ['b','c'],ascending=[True,False]))
```

```python
df.loc()#通过标签索引数据
df.iloc()#通过位置获取数据
```
