```python
from matplotlib import pyplot as plt
import random 
import matplotlib
from matplotlib import font_manager 
```

```python
x = range(0,120)
y = [random.randint(20,35) for i in range(120)]

#
my_font = font_manager.FontProperties(fname="C:\Windows\Fonts\simhei.ttf")



fig = plt.figure(figsize=(20,8),dpi=80)





_xticks_lables = ["{}".format(i) for i in range(60)]
_xticks_lables += ["{}".format(i) for i in range(60)]

plt.xticks(list(x)[::5],_xticks_lables[::5],rotation=45,fontproperties=my_font)   #rotation



# #
plt.xlabel("",fontproperties=my_font)
plt.ylabel("",fontproperties=my_font)
plt.title("",fontproperties=my_font)



#
plt.grid(alpha=0.3)#alpha


plt.plot(x,y)
plt.show()
```
