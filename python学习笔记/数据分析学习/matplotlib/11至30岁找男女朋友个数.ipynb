```python
from matplotlib import pyplot as plt
import random 
import matplotlib
from matplotlib import font_manager 
```

```python
#
my_font = font_manager.FontProperties(fname="C:\Windows\Fonts\simhei.ttf")

y_1 = [1,3,2,4,5,0,1,1,4,2,3,2,3,1,1,5,1,1,0,1]
y_2 = [1,1,2,1,5,2,2,0,1,2,1,0,5,1,1,2,1,3,2,1]
x = range(11,31)

fig = plt.figure(figsize=(20,8),dpi=80)

xtick_label = ["{}".format(i) for i in x]
plt.xticks(x,xtick_label,fontproperties=my_font)
plt.yticks(range(0,9))


#
plt.plot(x,y_1,label="")
plt.plot(x,y_2,label="")



#
plt.legend(prop=my_font)

plt.show()

```
