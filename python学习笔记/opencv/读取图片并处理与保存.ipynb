```python
#在开头引入必要的库
import matplotlib.pyplot as plt
import numpy as np
import cv2
from keras.preprocessing import image
from PIL import Image
%matplotlib inline
```

```python
#读取一张图片
xx = cv2.imread('D:/AV_Pro/2D_gray/Sample2_raw000.jpg',1)
print(xx.shape)
#img这一参数中填入图片的完全路径或者相对路径
cv2.imshow('image',xx)
cv2.waitKey(0)
```

    cv2.IMREAD_COLOR : 读入图片,任何与透明度相关通道的会被忽视,默认以这种方式读入.
    cv2.IMREAD_GRAYSCALE : 以灰度图的形式读入图片.
    cv2.IMREAD_UNCHANGED : 保留读取图片原有的颜色通道.
    可以简单的用-1,0,1来分别表示这3个flag

```python
#读取一张图片
xx = cv2.imread('D:/AV_Pro/2D_gray/Sample2_raw000.jpg',0)
xx = cv2.cvtColor(xx,cv2.COLOR_BGR2RGB)
print(xx.shape)
#img这一参数中填入图片的完全路径或者相对路径
plt.imshow(xx)
plt.axis("off")#去除坐标轴
plt.show()
```

```python
#读取并保存一张图片
xx = cv2.imread('D:/AV_Pro/2D_gray/Sample2_raw100.jpg')
xx = cv2.cvtColor(xx,cv2.COLOR_BGR2GRAY)
print(xx.shape)
#img这一参数中填入图片的完全路径或者相对路径
plt.imshow(xx)
plt.axis("off")#去除坐标轴
plt.show()

cv2.imwrite('D:/AV_Pro/1.jpg', xx,[cv2.IMWRITE_JPEG_QUALITY,100])#图片质量0-100

#jpg有损压缩0-100，png无损压缩0-9
```
